```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
hppdataset=pd.read_csv('housing_price_prediction_train.csv')
```

```python
hppdataset.head()
```

```python
print(hppdataset.shape)
```

```python
sns.heatmap(hppdataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#check the percentage of Nan vlaues present ineach faature
#finding the features which has missing values
features_with_missingval=[features for features in hppdataset.columns if hppdataset[features].isnull().sum()>1]

for feature in features_with_missingval:
    print(feature,np.round(hppdataset[feature].isnull().mean(),4), '% missing values ')
```

```python
#understanding the relationship between missiing value Feature and salesprice
for feature in features_with_missingval:
    data=hppdataset.copy()
   #COMMAND TO CONVERT NAN VALUE TO 1 AND OTHER TO ZEROS
    data[feature]=np.where(data[feature].isnull(),1,0)
    nd=data.groupby(feature)['SalePrice'].median().plot(kind='bar')
    #ax=nd.plot(kind='bar', figsize=(10,6), color="indigo", fontsize=13)
    print(nd)
    #plt.title(feature)
    plt.show()
    
```

```python
# IT IS OBSERVED THAT FEATURES WITH MISSING VALUES HAVE ALSO GOT SIGNIFICANCE 
#AS MEDIAN VALUES OF SALES PRICE WITH MISSING DATA FEATURE IS HIGHER
# EG. IN FEAtURE  MasVnrType
# 0    162700  MEDIAN VALUE WHERE DATA IS PRESENT
# 1    203287  MEDIAN VALUE WHER DATA IS ABSENT
#
```

```python
# Now listing the features with numerical data
num_feat=[features for features in hppdataset.columns if hppdataset[features].dtypes != 'O']
non_num_feat=[feature for feature in hppdataset.columns if hppdataset[feature].dtypes == 'O']
print(len(num_feat))
print(len(non_num_feat))
print(num_feat)
print(non_num_feat)             
              
```

```python
#train['Age'].hist(bins=30,color='darkred',alpha=0.3)
hppdataset['SalePrice'].hist(bins=30,color='darkred',alpha=0.3)
```

```python
feat_with_year= [feature for feature in num_feat if('Yr' in feature or 'Year' in feature)]
print(feat_with_year)                 
```

```python
hppdataset.groupby('YrSold')['SalePrice'].median().plot()
plt.show()
```

```python
#NOW WE WILL CHECK IF AGE OF HOUSE HAS ANY EFFECT ON SALE PRICE OR 
# OR RENOVATION YEAR HAS ANY EFFECT ON SALE PRICE
newdata1=hppdataset['YrSold']- hppdataset['YearBuilt']
plt.scatter(newdata1,hppdataset['SalePrice'])
plt.show()

newdata2=hppdataset['YrSold'] - hppdataset['YearRemodAdd']
plt.scatter(newdata2,hppdataset['SalePrice'])
plt.show()
```

```python
# ABOVE GRAPHS SHOWS THAT OLDER THE HOUSE LESSER IS THE PRICE
#ALSO NEWLY RENOVATED HOUSES ALSO SOLD AT HIGHER PRICE
```

```python
# nOW IDENTIFYING THE DISCRETE AND CONTINUOUS VARIABLES
dkt_feat=[feature for feature in num_feat if len(hppdataset[feature].unique())<25 and feature not in feat_with_year+['Id']]
cont_feat=[feature for feature in num_feat if feature not in dkt_feat+feat_with_year+['Id']]
print(dkt_feat)
print(cont_feat)
```

```python
#nOW LET SEE THE VARIATION OF SALES_PRICE
for feature in dkt_feat:
    data=hppdataset.copy()
    data.groupby(feature)['SalePrice'].median().plot(kind='bar')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#nOW LET SEE THE SCATTER PLOT OF CONTINOUS VARIABLE FEATURE WITH SALE PRICE
for feature in cont_feat:
    plt.scatter(hppdataset[feature],hppdataset['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#nOW LET SEE THE HISTOGRAM PLOT OF CONTINOUS VARIABLE FEATURE 
for feature in cont_feat:
    hppdataset[feature].hist(bins=20)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
```

```python
# Performing the logarthimic transoformation for continuous  value feature
for feature in cont_feat:
    newhppdataset=hppdataset.copy()
    if 0 in newhppdataset[feature].unique():
        pass
    else:
        newhppdataset[feature]=np.log(newhppdataset[feature])
        newhppdataset['salePrice']=np.log(newhppdataset['SalePrice'])
        plt.scatter(newhppdataset[feature],newhppdataset['SalePrice'])
        plt.show()
```

```python
# creating histogram after logarthimic transformation
for feature in cont_feat:
    newhppdataset=hppdataset.copy()
    if 0 in newhppdataset[feature].unique():
        pass
    else:
        newhppdataset[feature]=np.log(newhppdataset[feature])
        newhppdataset[feature].hist(bins=20)
        plt.show()
    
```

```python
#finding featurs which has max outliers using Boxplot
for feature in cont_feat:
    newhppdataset=hppdataset.copy()
    if 0 in newhppdataset[feature].unique():
        pass
    else:
        newhppdataset[feature]=np.log(newhppdataset[feature])
        newhppdataset.boxplot(column=feature)
        plt.show()
```
