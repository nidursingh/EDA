```python
#FEATURE ENGINEERING DONE FOR THE FEATURE WHICH HAS NAN VALUES
#FIRST FINDING THE CATEGORICAL FEATURES WHICH HAS NAN VALUE

cat_feat_with_missingval=[feature for feature in hppdataset.columns if hppdataset[feature].isnull().sum()>1 and hppdataset[feature].dtypes == 'O']
for feature in cat_feat_with_missingval:
    print('{}: {}% missing values'.format(feature,np.round(hppdataset[feature].isnull().mean(),4)))
```

```python
hppdataset[cat_feat_with_missingval].head()
```

```python
##NOW REPLACING MISSING VALUE WITH A NEW LABEL 'MISSING'
for feature in cat_feat_with_missingval:
    hppdataset[feature]=hppdataset[feature].fillna("Missing")
    hppdataset[feature].isnull().sum()
    
for feature in cat_feat_with_missingval:
    print('{}: {}% missing values'.format(feature,np.round(hppdataset[feature].isnull().mean(),4)))
```

```python
hppdataset[cat_feat_with_missingval].head()
```

```python
#FINDING THE FEATURE WHICH NUMERICA DATA AND NAN VALUES  
numeric_feat_missingval=[feature for feature in hppdataset.columns if hppdataset[feature].isnull().sum()>1 and hppdataset[feature].dtypes != 'O']

#FIND THE PERCENTAGE OF MISSING VALUE IN NUMERIC FEATURES
for feature in numeric_feat_missingval:
    print('{}: {}% missing value'.format(feature,np.round(hppdataset[feature].isnull().mean(),4)))
```

```python
hppdataset[numeric_feat_missingval].head()
```

```python
##REPLACING THE MISSING VALUE WITH MEDIAN VALUE AS THERE ARE LOT 
#OF OUTLIERS IN THE MISSING FEATURE VALUE. IF THERE ARE NOT OUTLIERS THEN
# REPLACING THE MISSING VALUE WITH MEAN WOULD HAVE BEEN A IDEAL CHOICE
for feature in numeric_feat_missingval:
    median_val=hppdataset[feature].median()
    hppdataset[feature+'nan']=np.where(hppdataset[feature].isnull(),1,0)
    hppdataset[feature].fillna(median_val,inplace=True)
hppdataset[numeric_feat_missingval].isnull().sum()    
```

```python
hppdataset[numeric_feat_missingval].head(100)
```

```python
#modifying the feature with year data to no of year built etc
#finding the feature with year
feat_with_year=[feature for feature in hppdataset.columns if  'Year' in feature or 'Yr' in feature ]
print(feat_with_year)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    hppdataset[feature] = hppdataset['YrSold'] - hppdataset[feature]

hppdataset.head()
hppdataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
sns.heatmap(hppdataset.isnull())
```

```python
# FINDING ALL THE FEATURES WITH NUMERIC DATA and DIVIDING IT INTO 
#DISCRETE AND CONTINUOUS

numeric_feat=[feature for feature in hppdataset.columns if hppdataset[feature].dtypes != 'O']
dkt_numeric_feat=[feature for feature in numeric_feat  if len(hppdataset[feature].unique())<25 and feature not in feat_with_year+['Id']]
cont_numeric_feat=[feature for feature in numeric_feat if feature not in dkt_numeric_feat+feat_with_year+['Id']]
print(dkt_numeric_feat)
print(cont_numeric_feat)
```

```python
#NOW SELECTING FEW FEATURES WHICH HAVE SKEWED DATA AND
#PERFORMING LOG TRANFORMATION
num_features=['LotFrontage', 'LotArea','1stFlrSF' ,'GrLivArea', 'SalePrice']

for feature in num_features:
    hppdataset[feature].hist(bins=20)
    #plt.scatter(hppdataset[feature],hppdataset['SalePrice'])
    #plt.title(feature)
    plt.show
```

```python
num_features=['LotFrontage', 'LotArea','1stFlrSF' ,'GrLivArea', 'SalePrice']

for feature in num_features:
    hppdataset[feature]=np.log(hppdataset[feature])
    hppdataset[feature].hist(bins=20)
    #plt.scatter(hppdataset[feature],hppdataset['SalePrice'])
    #plt.title(feature)
    plt.show
```

```python
#HANDLING THE RARE CATEGORICAL FEATURES i.e, A CATEGORY IN NON NUMERIC FEATURE
# IS OCCURING LESS THAN 1%
for feature in non_numeric_feat:
    cat_count_frac=hppdataset.groupby(feature)['SalePrice'].count()/len(hppdataset)
    cat_count_frac_norare=cat_count_frac[cat_count_frac>0.01].index
    hppdataset[feature]=np.where(hppdataset[feature].isin(cat_count_frac_norare),hppdataset[feature],'Rare_Var')
```

```python
#LABEL ENCODER FOR CONVERTING NON NUMBERIC CATEGORICAL FEATURE TO NUMERIC DATA
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for feature in non_numeric_feat:
    hppdataset[feature]=le.fit_transform(hppdataset[feature])
    
```
